# Stage 1.

Let `U_trunc` be the operator obtained by calculating the truncated unitary (i.e. with numeric_truncated unitary) and let `U_circ` be the operator obtained from the quantum circuit obtained by `direct_decomposition`. The objective of this section is to first verify that `U_trunc = U_circ` for a single beamsplitter, and then verify the same thing when an ancilla qubit is added.

```python
import numpy as np
import sys
from qiskit.visualization import plot_histogram
import interferometer as itf

from interferometer_circuits.numeric_truncated_unitaries import *
from interferometer_circuits.direct_decomposition import *
from interferometer_circuits.boson_sampling_probabilities import *
from interferometer_circuits.simulation import *
from random import random
from qiskit.quantum_info import Operator
```

## 1.0)
### Single beamsplitter (Pass)

```python
# Make single beamsplitter interferometer
I = itf.Interferometer()
theta_1 = 0.1 # np.pi/2 * random()
phi_1 = 0.1 # np.pi * random()

BS1 = itf.Beamsplitter(1, 2, theta=theta_1, phi=phi_1)
I.add_BS(BS1)
I.draw()
```

```python
U_BS = I.calculate_transformation() # Beamsplitter of the unitary

U_trunc = numeric_truncated_unitary(U_BS, 1)
qc = direct_decomposition(U_BS, 1)
U_circ = Operator(qc).data
assert(np.allclose(U_trunc, U_circ))
print("U_trunc == U_circ")
```

## 1.1) 
### Single beamsplitter with multiple photons (Pass)

```python
U_BS = I.calculate_transformation() # Beamsplitter of the unitary
U_trunc = numeric_truncated_unitary(U_BS, 3)

qc = direct_decomposition(U_BS, 3)
U_circ = Operator(qc).data
result = np.isclose(U_trunc, U_circ)
assert result.all()
```

How do the elements of `U_trunc` and `U_circ` differ? Is it a matter of phase or magnitude?

```python
different_entries = []
for i, row in enumerate(np.isclose(U_trunc, U_circ)):
    for j, item in enumerate(row):
        if item == False:
            different_entries.append((i,j))

for entry in different_entries:
    a = U_trunc[entry[0], entry[1]]
    b = U_circ[entry[0], entry[1]]
    assert(np.isclose(np.absolute(a), np.absolute(b)))

print("Entries equal up to phase")
```

```python
def complex_angle(z):
    m = np.absolute(z) # magnitude
    res = -1j * np.log(z/m)
    assert np.isclose(res.imag, 0)
    return res.real

# Test (passing):
# np.absolute(1+7j) * np.exp(1j * complex_angle(1+7j))

print("U_trunc", "U_circ")
for entry in different_entries:
    a = U_trunc[entry[0], entry[1]]
    b = U_circ[entry[0], entry[1]]
    print(entry, complex_angle(a), complex_angle(b))
```

^ The two phase terms sum up to give either zero or $\pi$...

```python
def zero_close_elements(matrix, threshold=1e-10):
    """
    Sets elements in the complex matrix to zero if their magnitude is close to zero.
    
    Parameters:
    matrix (np.ndarray): A 2D numpy array with complex numbers.
    threshold (float): A small threshold value to determine what is considered close to zero.
    
    Returns:
    np.ndarray: A matrix with elements close to zero set to exactly zero.
    """
    real_close_to_zero = np.abs(matrix.real) < threshold
    imag_close_to_zero = np.abs(matrix.imag) < threshold
    
    matrix.real[real_close_to_zero] = 0
    matrix.imag[imag_close_to_zero] = 0
    
    return matrix
```
