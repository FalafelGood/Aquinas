How does Clements interferometer decompose 50-50 beamsplitters?

```python
import numpy as np
import interferometer as itf
```

```python
# Test 1: Passing 
BS = itf.Beamsplitter(1,2,np.pi/4,0)
I = itf.Interferometer()
I.add_BS(BS)
U = I.calculate_transformation()
print(U)
```

```python
I = itf.square_decomposition(np.matrix([[1/np.sqrt(2),1/np.sqrt(2)],[1/np.sqrt(2),-1/np.sqrt(2)]]))
print(I.BS_list)
print(I.output_phases)
(I.draw())
```

```python
# Construct unitary in reverse
# Test 1: Passing 
BS1 = itf.Beamsplitter(1,2,0,np.pi)
BS2 = itf.Beamsplitter(1,2,np.pi/4,0)
I = itf.Interferometer()
I.add_BS(BS1)
I.add_BS(BS2)
U = I.calculate_transformation()
print(U)
```
